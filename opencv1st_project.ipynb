```
#i used haar cascade methord to detect the front face of the image, but haar cascase methord has approx 6000+ face slices so, better to perform face detection before using haar cascade because if we compare each sildes (6000+) then it will be difficult and time consuming,
```

```
     import numpy as np
     import cv2
     import matplotlib.pyplot as plt
     
     face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
     
     
     img = cv2.imread('Capture.jpg')
     gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
     faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    roigray = gray[y:y+h, x:x+w]
    roicolor = img[y:y+h, x:x+w]
plt.grid(None)   
plt.xticks([])
plt.yticks([])
imgplot = plt.imshow(img)
```
